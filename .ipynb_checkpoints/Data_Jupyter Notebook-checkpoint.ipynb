## Cleansing data

```python
daftar_alay = pd.read_csv('new_kamusalay.csv', encoding='ISO-8859-1', header=None)
```

```python
daftar_alay.head(5)
```

```python
#ganti nama kolom 
daftar_alay = daftar_alay.rename(columns={0: 'awal', 1: 'ganti'})
```

```python
daftar_alay.head(5)
```

```python
#bersihkan karakter-karakter yang tidak perlu

def bersih_bersih(text):
    text = re.sub('\n',' ',text)
    text = re.sub('rt',' ',text)
    text = re.sub('user',' ',text)
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ',text)
    text = re.sub('  +', ' ', text)
    text = re.sub('[^0-9a-zA-Z]+', ' ', text)
    return text
```

```python
#jadikan teks kecil semua
def lowercase(text):
    return text.lower()
```

```python
#ganti kata-kata alay dari kamus alay yang tersedia
kamus_alay = dict(zip(daftar_alay['awal'], daftar_alay['ganti']))
def ganti_alay(text):
    return ' '.join([kamus_alay[word] if word in kamus_alay else word for word in text.split(' ')])
```

```python
#Jadikan satu fungsi cleansing teks
def cleaning(text):
    text = bersih_bersih(text) 
    text = lowercase(text)
    text = ganti_alay(text)
    return text
```

```python
cleaned_text = [cleaning(text) for text in tweets['Tweet']]
```

```python
tweets['cleaned_text'] = cleaned_text
```

## Labeling Data

```python
# Membaca file CSV yang berisi daftar kata-kata abusive
df_abusive = pd.read_csv('abusive.csv', header=None, names=['abusive_word'])
```

```python
df_abusive.head()
```

```python
# Mengonversi daftar kata-kata menjadi set untuk pencarian yang lebih efisien
set_abusive_words = set(df_abusive['abusive_word'])
```

```python
def label_data(text):
    words = text.split()
    if any(word in set_abusive_words for word in words):
        return "abusive"
    else:
        return "non-abusive"
```

```python
tweets['label'] = tweets['cleaned_text'].apply(label_data)
```

```python
tweets.head()
```

#### Quartile dan Interquartile Range

```python
# Fungsi untuk menghitung quartile dan interquartile range
def calculate_quartiles_and_iqr(column_values):
    q1 = np.percentile(column_values, 25)
    q2 = np.percentile(column_values, 50)
    q3 = np.percentile(column_values, 75)
    iqr = q3 - q1
    return q1, q2, q3, iqr
```

```python
# Menghitung quartile dan interquartile range untuk kolom 'total_char'
q1_char, q2_char, q3_char, iqr_char = calculate_quartiles_and_iqr(tweets['total_char'])
```

```python
# Menghitung quartile dan interquartile range untuk kolom 'total_word'
q1_word, q2_word, q3_word, iqr_word = calculate_quartiles_and_iqr(tweets['total_word'])
```

```python
# Menampilkan hasil
print(f"Quartiles and IQR for 'total_char':")
print(f"Q1: {q1_char}")
print(f"Q2: {q2_char}")
print(f"Q3: {q3_char}")
print(f"IQR: {iqr_char}")

print("\n")

print(f"Quartiles and IQR for 'total_word':")
print(f"Q1: {q1_word}")
print(f"Q2: {q2_word}")
print(f"Q3: {q3_word}")
print(f"IQR: {iqr_word}")
```

```python
# Menentukan batas atas dan batas bawah untuk 'total_char'
upper_bound_char = q3_char + 1.5 * iqr_char
lower_bound_char = q1_char - 1.5 * iqr_char

# Menentukan batas atas dan batas bawah untuk 'total_word'
upper_bound_word = q3_word + 1.5 * iqr_word
lower_bound_word = q1_word - 1.5 * iqr_word
```

```python
# Menentukan outlier untuk 'total_char'
outliers_char = tweets[(tweets['total_char'] > upper_bound_char) | (tweets['total_char'] < lower_bound_char)]

# Menentukan outlier untuk 'total_word'
outliers_word = tweets[(tweets['total_word'] > upper_bound_word) | (tweets['total_word'] < lower_bound_word)]
```

```python
# Menghitung jumlah outliers
num_outliers_char = outliers_char.shape[0]
num_outliers_word = outliers_word.shape[0]
```

```python
# Menampilkan hasil
print(f"Batas atas dan batas bawah untuk 'total_char':")
print(f"Batas atas: {upper_bound_char}")
print(f"Batas bawah: {lower_bound_char}")
print(f"Jumlah outliers untuk 'total_char': {num_outliers_char}")
print(f"Outliers: \n{outliers_char}")

print("\n")

print(f"Batas atas dan batas bawah untuk 'total_word':")
print(f"Batas atas: {upper_bound_word}")
print(f"Batas bawah: {lower_bound_word}")
print(f"Jumlah outliers untuk 'total_word': {num_outliers_word}")
print(f"Outliers: \n{outliers_word}")
```

#### Menghitung kata abusive terbanyak yang muncul

```python
abusive_words = pd.read_csv('abusive.csv', header=None)
abusive_words = abusive_words[0].tolist()

# Inisialisasi kamus untuk menghitung frekuensi kata-kata
word_freq = {}
```

```python
# Loop melalui setiap teks dalam data yang telah dibersihkan
for text in tweets['cleaned_text']:
    words = text.split()
    for word in words:
        if word in abusive_words:
            if word in word_freq:
                word_freq[word] += 1
            else:
                word_freq[word] = 1
```

```python
# Ambil 10 kata terbanyak
top_abusive_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:10]
```

```python
# Tampilkan 10 kata terbanyak
for word, freq in top_abusive_words:
    print(f"Kata: {word}, Frekuensi: {freq}")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
```

#### Menghitung 10 kata yang sering muncul untuk semua label

```python


# Menggunakan nltk untuk tokenisasi dan menghapus stopwords
stop_words = set(stopwords.words('indonesian'))

def process_text(text):
    tokens = word_tokenize(text)
    words = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
    return words

# Menggabungkan semua kata dari kolom 'text'
all_words = tweets['text_remove_stopwords'].apply(process_text).explode()

# Menghitung frekuensi kata-kata
word_counts = Counter(all_words)

# Menampilkan 10 kata terbanyak
top_10_words = word_counts.most_common(10)
print("10 kata terbanyak:")
for word, count in top_10_words:
    print(f"{word}: {count}")
```
